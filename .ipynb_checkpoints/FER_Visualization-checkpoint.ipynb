```python
import os
import PIL
import numpy as np
import torch
import torch.nn.functional as F
import torchvision.models as models
from torchvision import transforms
from torchvision.utils import make_grid, save_image
from torch import nn
from gradcam.utils import visualize_cam
from gradcam import GradCAM, GradCAMpp
import pretrainedmodels
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
url = "https://image.freepik.com/free-photo/close-up-happy-face-boy_1150-12122.jpg"
```

```python
from PIL import Image
import requests
from io import BytesIO
def im_process(url):

    ext = url.split(".")[1].lower()
    response = requests.get(url)
    file_name = BytesIO(response.content)
    im = Image.open(file_name)


    return im 
pil_img = im_process(url)
```

```python
pil_img
```

```python
torch_img = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor()
])(pil_img).to(device)
```

```python
normed_torch_img = (torch_img)[None]
```

```python
model_name = 'densenet121'
num_classes = 7
model_fn = pretrainedmodels.__dict__[model_name]
model = model_fn(num_classes=1000).to(device)
model.last_linear = nn.Linear(1024, num_classes).to(device)
```

```python
model_path='logs/trial/checkpoints/last.pth'
load = torch.load(model_path, map_location=torch.device('cuda'))["model_state_dict"]
model.eval()
```

```python
configs = [
    dict(model_type='densenet', arch=model, layer_name='features_norm5'),
]
```

```python
for config in configs:
    config['arch'].to(device).eval()
cams = [
    [cls.from_config(**config) for cls in (GradCAM, GradCAMpp)]
    for config in configs
]
```

```python
images = []
images2 = []
for gradcam, gradcam_pp in cams:
    mask, _ = gradcam(normed_torch_img)
    heatmap, result = visualize_cam(mask, torch_img)
    mask_pp, _ = gradcam_pp(normed_torch_img)
    heatmap_pp, result_pp = visualize_cam(mask_pp, torch_img)

    images.extend([torch_img.cpu(), heatmap, heatmap_pp, result, result_pp])
    images2.extend([torch_img.cpu(), result, result_pp])

grid_image = make_grid(images2, nrow=3)
```

```python
transforms.ToPILImage()(grid_image)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
y_pred=model(normed_torch_img).cpu().detach().numpy()
y_pred=sigmoid(y_pred)
y_pred
```
